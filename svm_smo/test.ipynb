```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs


data = pd.read_csv('cod-rna_data.txt', sep=" ", header=None)
label = pd.read_csv('cod-rna_label.txt', sep=" ", header=None)
label.head()
```

```python
sample_label = label.groupby(0, group_keys=False).apply(lambda x: x.sample(1000))
```

```python
sample_data = data.iloc[sample_label.index]
```

```python
sample_data = sample_data[[0, 1]]
sample_data
```

```python
sample_data.to_csv("sample_rna.txt", sep='\t', header = False, index = False)
```

```python
sample_label.to_csv("sample_rna_label.txt", sep='\t',  header = False, index = False)
```

```python
data, label = make_blobs(n_samples=3500, centers=2,
                        n_features=2, random_state=1)
data = pd.DataFrame(data)
label = pd.DataFrame(label)
```

```python
sample_label = label.groupby(0, group_keys=False).apply(lambda x: x.sample(1500))
```

```python
sample_data = data.iloc[sample_label.index]
test_data = data.iloc[np.logical_not(data.index.isin(sample_data.index))]
test_label = label.iloc[np.logical_not(label.index.isin(sample_label.index))]
```

```python
test_label.head()
```

```python
test_data.head()
```

```python
sample_data.to_csv("sample_data.txt", sep=' ', header = False, index = False)
sample_label.to_csv("sample_label.txt", sep=' ', header = False, index = False)
test_data.to_csv("test_data.txt", sep=' ', header = False, index = False)
test_label.to_csv("test_label.txt", sep=' ', header = False, index = False)
```
